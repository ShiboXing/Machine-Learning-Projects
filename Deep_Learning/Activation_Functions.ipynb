```python
%matplotlib inline
from matplotlib import pyplot as plt
from mxnet import nd, autograd

def xyplot(x_vals, y_vals, name):
    plt.figure(figsize=(10, 5))
    plt.scatter(x_vals, y_vals)
    axes = plt.gcf().gca()
    plt.axis([-10, 10, -2, 2])
    plt.xlabel('x')
    plt.ylabel(f'{name}(y)')
    plt.title(name)
    plt.show()
```

### relu


```python

x = nd.arange(-8, 8, 0.1, dtype=float)
x.attach_grad()
with autograd.record():
    y = x.relu()
xyplot(x.asnumpy(), y.asnumpy(), 'relu')
```

```python
y.backward()
xyplot(x.asnumpy(), x.grad.asnumpy(), 'relu_grad')
```

### sigmoid

```python
with autograd.record():
    y = x.sigmoid()
xyplot(x.asnumpy(), y.asnumpy(), 'sigmoid')

y.backward()
xyplot(x.asnumpy(), x.grad.asnumpy(), 'sigmoid_grad')
```

### tanh


```python
with autograd.record():
    y = x.tanh()
xyplot(x.asnumpy(), y.asnumpy(), 'tahn')
```

```python
y.backward()
xyplot(x.asnumpy(), x.grad.asnumpy(), 'tanh_grad')
```
